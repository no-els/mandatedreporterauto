```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
import re
import time

# Function to extract principals and assistant principals from a given HTML content
def extract_principals(html_content):
    # Use regex to find elements containing "Principal"
    principal_pattern = re.compile(r'<li class="field-item">.*?<h3 class="field-item_name">Principal</h3>.*?<div class="field-item_value"><ul><li>(.*?)</li></ul></div>', re.DOTALL)
    principal_matches = principal_pattern.findall(html_content)

    # Use regex to find elements containing "Assistant Principal"
    assistant_principal_pattern = re.compile(r'<li class="field-item">.*?<h3 class="field-item_name">Assistant Principal\(s\)?</h3>.*?<div class="field-item_value"><ul><li>(.*?)</li></ul></div>', re.DOTALL)
    assistant_principal_matches = assistant_principal_pattern.findall(html_content)

    # Extract and categorize the names
    principals = [match.strip() for match in principal_matches]
    assistant_principals = [match.strip() for match in assistant_principal_matches]

    return principals, assistant_principals

# Set up Chrome options
chrome_options = Options()
chrome_prefs = {
    "download.default_directory": "/Users/noelngui/Code/mission_graduates_eleyo_lib/contact_files",  # Set download path
    "download.prompt_for_download": False,
    "profile.default_content_settings.popups": 0,
    "safebrowsing.enabled": True
}
chrome_options.add_experimental_option("prefs", chrome_prefs)

# Initialize the WebDriver
driver = webdriver.Chrome(options=chrome_options)

# Read the URLs from websites.txt
with open('websites.txt', 'r') as file:
    urls = file.readlines()

# Process each URL
for url in urls:
    url = url.strip()
    if url:
        print(f"Processing {url}...")
        
        # Open the URL
        driver.get(url)

        # Allow the page to load
        time.sleep(2)

        # Extract the HTML content of the page
        html_content = driver.page_source

        # Extract principals and assistant principals
        principals, assistant_principals = extract_principals(html_content)

        # Print the categorized results
        if principals:
            print("Principals:")
            for principal in principals:
                print(f" - {principal}")
        
        if assistant_principals:
            print("Assistant Principals:")
            for assistant_principal in assistant_principals:
                print(f" - {assistant_principal}")
        
        print("\n")

# Close the WebDriver
driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
import time
import csv
import re

# Function to extract email and name from a given HTML content
def extract_email_and_name(driver):
    try:
        # Check if the page contains the "Sorry, the page you are looking for cannot be found." message
        if "Sorry, the page you are looking for cannot be found." in driver.page_source:
            return None, None, False
        
        # Extract the name
        name_element = driver.find_element(By.CSS_SELECTOR, 'h1')
        name = name_element.text.strip()

        # Extract the email from the specified section
        content_element = driver.find_element(By.CSS_SELECTOR, '#block-sfusd-base-content > article > div')
        content_text = content_element.text

        # Use regex to find the email address
        email_match = re.search(r'\b[A-Za-z0-9._%+-]+@sfusd\.edu\b', content_text)
        if email_match:
            email = email_match.group(0)
            return name, email, True
        else:
            return name, None, True
    except Exception as e:
        return None, None, False

# Set up Chrome options
chrome_options = Options()
chrome_prefs = {
    "download.default_directory": "/Users/noelngui/Code/mission_graduates_eleyo_lib/contact_files",  # Set download path
    "download.prompt_for_download": False,
    "profile.default_content_settings.popups": 0,
    "safebrowsing.enabled": True
}
chrome_options.add_experimental_option("prefs", chrome_prefs)

# Initialize the WebDriver
driver = webdriver.Chrome(options=chrome_options)

# Read the URLs from people.txt
with open('people.txt', 'r') as file:
    urls = file.readlines()

# Open the CSV file for writing
with open('people_emails.csv', 'w', newline='') as csvfile:
    fieldnames = ['URL', 'Name', 'Email', 'Site Exists', 'Email Exists']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()

    # Process each URL
    for url in urls:
        url = url.strip()
        if url:
            print(f"Processing {url}...")
            
            # Open the URL
            driver.get(url)

            # Allow the page to load
            time.sleep(2)

            # Extract the email and name
            name, email, site_exists = extract_email_and_name(driver)

            # Determine if the email exists
            email_exists = bool(email)

            # Write the extracted data to the CSV file
            writer.writerow({
                'URL': url,
                'Name': name if name else 'N/A',
                'Email': email if email else 'N/A',
                'Site Exists': 'Yes' if site_exists else 'No',
                'Email Exists': 'Yes' if email_exists else 'No'
            })

            # Flush the buffer to ensure data is written to the file
            csvfile.flush()

# Close the WebDriver
driver.quit()

print("Data has been written to people_emails.csv")
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
import time
import csv

# Function to check if the specified <h3> element appears on the webpage and grab the website URL if possible
def check_website_h3_and_grab_url(driver):
    try:
        # Click the specified element first
        summary_element = driver.find_element(By.CSS_SELECTOR, '#block-sfusd-base-content > article > div > div.section-school-tabs.is-section-aligned-- > div > div.section-school-tabs_grid > div.section-school-tabs_content > div > details:nth-child(4) > summary')
        summary_element.click()
        time.sleep(1)  # Allow time for the content to load

        # Check if the <h3> element with class 'field-item_name' and text 'Website' appears
        h3_elements = driver.find_elements(By.CSS_SELECTOR, 'h3.field-item_name')
        for h3 in h3_elements:
            if h3.text.strip() == "Website":
                # Look for the URL within the specified selector
                details_div = driver.find_element(By.CSS_SELECTOR, '#block-sfusd-base-content > article > div > div.section-school-tabs.is-section-aligned-- > div > div.section-school-tabs_grid > div.section-school-tabs_content > div > details:nth-child(4) > div')
                links = details_div.find_elements(By.TAG_NAME, 'a')
                for link in links:
                    website_url = link.get_attribute('href')
                    if website_url:
                        return True, website_url
                return True, None
        return False, None
    except Exception as e:
        return False, None

# Set up Chrome options
chrome_options = Options()
chrome_prefs = {
    "download.default_directory": "/Users/noelngui/Code/mission_graduates_eleyo_lib/contact_files",  # Set download path
    "download.prompt_for_download": False,
    "profile.default_content_settings.popups": 0,
    "safebrowsing.enabled": True
}
chrome_options.add_experimental_option("prefs", chrome_prefs)

# Initialize the WebDriver
driver = webdriver.Chrome(options=chrome_options)

# Read the URLs from websites.txt
with open('websites.txt', 'r') as file:
    urls = file.readlines()

# Open the CSV file for writing
with open('website_h3_urls.csv', 'w', newline='') as csvfile:
    fieldnames = ['URL', 'Website H3 Exists', 'Website URL']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()

    # Process each URL
    for url in urls:
        url = url.strip()
        if url:
            print(f"Processing {url}...")
            
            # Open the URL
            driver.get(url)

            # Allow the page to load
            time.sleep(2)

            # Check if the specified <h3> element appears on the webpage and grab the website URL if possible
            website_h3_exists, website_url = check_website_h3_and_grab_url(driver)

            # Write the result to the CSV file
            writer.writerow({
                'URL': url,
                'Website H3 Exists': 'Yes' if website_h3_exists else 'No',
                'Website URL': website_url if website_url else 'N/A'
            })

            # Flush the buffer to ensure data is written to the file
            csvfile.flush()

# Close the WebDriver
driver.quit()

print("Data has been written to website_h3_urls.csv")
```
